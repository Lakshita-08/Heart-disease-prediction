```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
dataset = pd.read_csv('../input/heart-disease-dataset/heart.csv')
dataset.head()
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
X = dataset.drop(['target'],axis=1)
y = dataset.iloc[:, -1].values
```


**Data preprocessing**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Logistic Regression

```python
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
accuracy=accuracy_score(y_test,y_pred)
accuracy
```

# K-Nearest Neighbors (K-NN)

```python
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

# SVM

```python
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred= classifier.predict(X_test)
y_pred
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

# Decision Tree Classification

```python
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0, max_depth = 6)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

# Random Forest Classification

```python
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0, max_depth = 6)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```
